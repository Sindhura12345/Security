```python
import pandas as pd
import requests
import time
from google.colab import files

# Function to scan domain using VirusTotal API
def check_domain(domain, api_key):
    url = f"https://www.virustotal.com/api/v3/domains/{domain}"
    headers = {
        "accept": "application/json",
        "x-apikey": api_key
    }

    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            data = response.json()
            stats = data['data']['attributes']['last_analysis_stats']
            return {
                'domain': domain,
                'malicious': stats['malicious'],
                'suspicious': stats['suspicious'],
                'clean': stats['harmless'],
                'undetected': stats['undetected']
            }
        else:
            return {
                'domain': domain,
                'malicious': 'Error',
                'suspicious': 'Error',
                'clean': 'Error',
                'undetected': 'Error'
            }
    except Exception as e:
        print(f"Error scanning {domain}: {str(e)}")
        return {
            'domain': domain,
            'malicious': 'Error',
            'suspicious': 'Error',
            'clean': 'Error',
            'undetected': 'Error'
        }

# Main execution
def main():
    # Upload Excel file
    print("Please upload your Excel file containing domains...")
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]

    # Read API key
    api_key = input("Enter your VirusTotal API key: ")

    # Read domains from Excel
    df = pd.read_excel(filename)
    domain_column = input("Enter the name of column containing domains: ")

    results = []
    total_domains = len(df[domain_column])

    print(f"\nStarting scan of {total_domains} domains...")

    # Process each domain
    for index, domain in enumerate(df[domain_column], 1):
        print(f"Scanning {index}/{total_domains}: {domain}")
        result = check_domain(domain, api_key)
        results.append(result)
        time.sleep(15)  # Respect API rate limits

    # Create results DataFrame
    results_df = pd.DataFrame(results)

    # Save results
    output_filename = 'virustotal_results.xlsx'
    results_df.to_excel(output_filename, index=False)
    files.download(output_filename)

    print("\nScan complete! Results saved to virustotal_results.xlsx")

if __name__ == "__main__":
    main()
```
